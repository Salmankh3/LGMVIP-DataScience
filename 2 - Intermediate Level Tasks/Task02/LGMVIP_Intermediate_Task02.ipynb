<b>Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn import tree
```

<b>Loading Dataset

```python
data = pd.read_csv("iris.data.csv")
```

<b>Making X And Y Sets From Data

```python
X_list = ['Sepal Length','Sepal Width', 'Petal Length', 'Petal Width']
X = data[X_list]
Y = data['Specie']
```

```python
X.head(10)
```

```python
Y.head(10)
```

<b>Encoding Categorical Target Feature (Y)

```python
encoder = LabelEncoder()
Y = encoder.fit_transform(Y)
print(Y)
```

<b>Making Train And Test Sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=3, train_size=0.8)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

<b>Creating Decision Tree Model And Fitting Train Values

```python
model = DecisionTreeClassifier()

model.fit(X_train,Y_train)
```

<b>Predicting Test Sets

```python
Y_pred = model.predict(X_test)
```

```python
print("Original Values \n {}".format(Y_test))
print("Test Predictions \n {}".format(Y_pred))
```

<b>Model Evaluations And Report

```python
print("Accuracy of Decision Tree Classifier: ", round(model.score(X_test,Y_test)*100,2))
```

```python
print(classification_report(Y_test,Y_pred))
```

```python
Y_list = ['Setosa', 'Versicolor', 'Virginica']
fig, axes = plt.subplots(nrows = 1, ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(model, feature_names = X_list, class_names=Y_list, filled = True)
plt.title('Our Model Tree Diagram', fontsize=5)
plt.show()
```
