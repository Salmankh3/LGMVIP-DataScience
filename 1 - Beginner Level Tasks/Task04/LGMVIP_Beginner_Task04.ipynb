<b>Importing Packages

```python
import cv2
import matplotlib.pyplot as plt
```

<b>Reading Image

```python
img = cv2.imread('Danny_Zee.PNG')

# img.shape
plt.figure(figsize=(8,8))
plt.imshow(img)
```

<b>Converting From BGR to RGB

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(RGB_img)
```

<b>Converting Into Grayscale

```python
grayscale = cv2.cvtColor(RGB_img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(grayscale)
```

<b>Inverting The Grayscale Image

```python
inverted = cv2.bitwise_not(grayscale)

plt.figure(figsize=(8,8))
plt.imshow(inverted)
```

<b>Blurring The Inverted Image

```python
blurred = cv2.GaussianBlur(inverted, (21,21), 0)

plt.figure(figsize=(8,8))
plt.imshow(blurred)
```

<b>Inverting The Blurred Image

```python
inverted_blurred = cv2.bitwise_not(blurred)

plt.figure(figsize=(8,8))
plt.imshow(inverted_blurred)
```

<b>Dividing Grayscale Image by Inverted-Blurred Image

```python
final = cv2.divide(grayscale, inverted_blurred, scale=256.0)

plt.figure(figsize=(8,8))
plt.imshow(final)
```

<b>The Final Pencil Sketch 

```python
final = cv2.cvtColor(final, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8,8))
plt.imshow(final)
```

<b>Saving The Pencil Sketch Image

```python
cv2.imwrite('output.PNG', final)
```
